## gensim word2vec, fasttext 파라미터

size = 워드 벡터의 특징 값. 즉, 임베딩 된 벡터의 차원.
- [50, 100]


window = 컨텍스트 윈도우 크기
- [3, 4, 5]


min_count = 단어 최소 빈도 수 제한 (빈도가 적은 단어들은 학습하지 않는다.)
- 50


workers = 학습을 위한 프로세스 수 
- default


sg = 0은 CBOW, 1은 Skip-gram.
- [0또는 1]

#### 복합성 키워드
1. 보습, 보습력,수분, 수분감 (보습감, 수분감)
2. 유분기, 기름 (유분기)
3. 트러블, 진정, 여드름, 자극, 재생크림 (피부 진정)
4. 가격, 용량 (가성비)
5. 주름, 탄력 (주름)
6. 미백, 기미, 톤업 (톤업)

```python
# 다정: 복합성 word2vec
# 1글자, 불용어 지움

# 불용어 리스트 _ ranks nl 한국어 불용어 사전 텍스트파일 불러오기
with open("stopwords_kor.txt", 'r') as file:
    lines = file.readlines()
    stopwords = [line.rstrip('\n') for line in lines]

# 불용어, 1글자 제거
clean_n = []
token_n = list(data[data['skin_type']=='복합성']['token_noun'].values)
for i in range(len(token_n)):
    clean_n.append([word for word in token_n[i] if word not in stopwords and len(word)>1])

print(len(data[data['skin_type']=='복합성']))
print(len(token_n))
```

```python
from gensim.models import Word2Vec

word2vec_bok = []

for size in [50, 100]:
    for window in [3, 4, 5]:
        for sg in [0, 1]:
            model = Word2Vec(sentences=clean_n, vector_size=size, window=window,
                             min_count=50, sg=sg)
            dictionary = {"model":model, "size":size, "window":window, "sg":sg}
            word2vec_bok.append(dictionary)
            
for i in range(len(word2vec_bok)):
    model = word2vec_bok[i]['model']
    print(i, 'st model: ')
    print(model.wv.most_similar("수분"), '\n')
```

```python
# 합성명사 합치기 (사용자 사전 수정 후)

#trigram
tri_remove = ['블레미쉬', '차앤박', '리페어크림', '리뉴얼', '아벤느',
                '리얼베리어', '바이오힐보', '라로슈포', '아토베리어',
                '바이오더마', '보타닉', '닥터자르카', '베리어익스트림',
               '자작나무수분크림', '닥터디런', '자르카페어']
list_3gram = []

for sent in token_noun:
    i=0
    while(i < len(sent)-2):
    
        if sent[i]+sent[i+1]+sent[i+2] == "피부진정":
            sent[i]="진정"
            del sent[i+1]
            del sent[i+1]
        elif sent[i]+sent[i+1]+sent[i+2] == "예민피부":
            sent[i]="예민"
            del sent[i+1]
            del sent[i+1]
        elif sent[i]+sent[i+1]+sent[i+2] == "피부진정효과":
            sent[i]="진정"
            del sent[i+1]
            del sent[i+1]        
        # 나머지 합성명사는 합쳐서 삭제
        elif sent[i]+sent[i+1]+sent[i+2] in tri_remove:
            del sent[i]
            del sent[i]
            del sent[i]
        
        i += 1
        
    list_3gram.append(sent)
    
# bigram
bi_remove = ['에스트라', '올리브영', '피지오', '바이오힐', '독도토너',
             '멀티밤']

list_2gram = []

for sent in list_3gram:
    i=0
    while(i < len(sent)-1):
        
        if sent[i]+sent[i+1] == "수분부족지성":
            sent[i]="수부지" # 원래 3gram에 있었으나 '수분부족'이 하나로 합쳐져서
            del sent[i+1]
        elif str(sent[i]+sent[i+1]) == "토너팩":
            sent[i]="스킨팩"
            del sent[i+1]
        elif str(sent[i]+sent[i+1]) == "강추":
            sent[i]="강추"
            del sent[i+1]
        # 나머지 합성명사는 합쳐서 삭제
        elif sent[i]+sent[i+1] in bi_remove:
            del sent[i]
            del sent[i]
            
        i += 1
        
    list_2gram.append(sent)

data['token_noun']=list_2gram
```

```python
# 불용어 리스트 _ ranks nl 한국어 불용어 사전 텍스트파일 불러오기
with open("stopwords_kor.txt", 'r') as file:
    lines = file.readlines()
    stopwords = [line.rstrip('\n') for line in lines]

# 불용어, 1글자 제거
clean_n = []
token_n = list(data[data['skin_type']=='복합성']['token_noun'].values)
for i in range(len(token_n)):
    clean_n.append([word for word in token_n[i] if word not in stopwords and len(word)>1])

print(len(data[data['skin_type']=='복합성']))
print(len(token_n))
```

```python
from gensim.models import Word2Vec

word2vec_bok = []

for size in [50, 100]:
    for window in [3, 4, 5]:
        for sg in [0, 1]:
            model = Word2Vec(sentences=clean_n, vector_size=size, window=window,
                             min_count=50, sg=sg)
            dictionary = {"model":model, "size":size, "window":window, "sg":sg}
            word2vec_bok.append(dictionary)
            
# for i in range(len(word2vec_bok)):
#     model = word2vec_bok[i]['model']
#     print(i, 'st model: ')
#     print(model.wv.most_similar("수분"), '\n')
```

```python
from gensim.models import FastText

fasttext_bok = []

for size in [50, 100]:
    for window in [3, 4, 5]:
        for sg in [0, 1]:
            model = FastText(clean_n, vector_size=size, window=window,
                             min_count=50, sg=sg)
            dictionary = {"model":model, "size":size, "window":window, "sg":sg}
            fasttext_bok.append(dictionary)
```

```python
import collections

keywords = [['보습', '보습력', '수분', '수분감'],
['유분기', '기름'],
['트러블', '진정', '여드름', '자극', '재생크림'],
['가격', '용량'],
['주름', '탄력'],
['미백', '기미', '톤업']]

most_similar_wv = [] # 키워드마다
most_similar_ft = []

for topic in keywords:
    for key in topic:
        similar_wv = []
        similar_ft = []
        for i in range(len(word2vec_bok)):
            
            model = word2vec_bok[i]['model']
            for word, _ in model.wv.most_similar(key):
                similar_wv.append(word)
                
            model = fasttext_bok[i]['model']
            for word, _ in model.wv.most_similar(key):
                similar_ft.append(word)
                
        count_wv = collections.Counter(similar_wv)
        count_ft = collections.Counter(similar_ft)
        most_similar_wv.append(count_wv.most_common())
        most_similar_ft.append(count_ft.most_common())
```

```python
embedding_wv = pd.DataFrame()
embedding_wv['skin_type'] = ['복합성']*18
embedding_wv['keyword'] = sum(keywords, [])
embedding_wv['most_similar'] = most_similar_wv
embedding_wv
```

```python
embedding_ft = pd.DataFrame()
embedding_ft['skin_type'] = ['복합성']*18
embedding_ft['keyword'] = sum(keywords, [])
embedding_ft['most_similar'] = most_similar_ft
embedding_ft
```
