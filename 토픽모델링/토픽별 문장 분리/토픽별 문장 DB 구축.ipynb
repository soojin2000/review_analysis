```python
import pandas as pd

oily = pd.read_csv('oily.csv', encoding='cp949')
oily.drop(['Unnamed: 0', 'skin_type', 'split_sentence'], axis=1, inplace=True)
oily
```

```python
from ast import literal_eval
oily['aspect_split'] = oily['aspect_split'].apply(literal_eval)
```

```python
# 키워드 -> 리스트로 저장

with open('oily_keyword.txt', 'r') as f:
    lines = f.read().splitlines()
    oily_topic1 = lines[0].split(',') # 트러블 진정
    oily_topic2 = lines[1].split(',') # 가성비
    oily_topic3 = lines[2].split(',') # 수분/보습
    oily_topic4 = lines[3].split(',') # 피부개선
    
print(oily_topic1)
print(oily_topic2)
print(oily_topic3)
print(oily_topic4)
```

```python
# 키워드 포함 -> 속성 분류 함수
def is_aspect_oily(sent):
    asp = [] # 1개 이상일 수 있음
    
    for word, tag in sent:
        if word in oily_topic1:
            asp.append('트러블 진정')
            break
            
    for word, tag in sent:
        if word in oily_topic2:
            asp.append('가성비')
            break
    
    for word, tag in sent:
        if word in oily_topic3:
            asp.append('수분/보습')
            break
            
    for word, tag in sent:
        if word in oily_topic4:
            asp.append('피부개선')
            break
    
    return asp
```

```python
# aspect_split(2차원) -> 1차원 data로 df형태 변형

rev_product = [] # 리뷰 포함된 제품
rev_rating = [] # 리뷰 별점
rev_origin = [] # 원본 리뷰
rev_split = [] # 리뷰 내 문장 1개
rev_aspect = [] # 문장이 포함될 속성

#omit_idx = []

for i in range(len(oily)): # 리뷰 하나 들어감
    for sent in oily.iloc[i]['aspect_split']: # 문장 하나씩
        belong = is_aspect_oily(sent) # 포함되는 속성->리스트로 반환
        
        if len(belong)==0: # 문장 내 키워드 존재 x
            rev_aspect.append('키워드 없음')
            rev_split.append(sent)
            rev_origin.append(oily.iloc[i]['review'])
            rev_rating.append(oily.iloc[i]['rating'])
            rev_product.append(oily.iloc[i]['product_name'])
        
        else:
            for j in range(len(belong)): # 1개 이상의 서로 다른 속성
                rev_aspect.append(belong[j])
                rev_split.append(sent)
                rev_origin.append(oily.iloc[i]['review'])
                rev_rating.append(oily.iloc[i]['rating'])
                rev_product.append(oily.iloc[i]['product_name'])

                
oily_sentence = pd.DataFrame({'aspect':rev_aspect,
                              'sentence':rev_split,
                              'review':rev_origin,
                              'rating':rev_rating,
                              'product_name':rev_product})

oily_sentence
```

```python
oily_sentence_aspect = oily_sentence[oily_sentence['aspect']!='키워드 없음']
oily_sentence_aspect
```

```python
oily_sentence_aspect.to_csv('oily_sentence_aspect.csv')
```
