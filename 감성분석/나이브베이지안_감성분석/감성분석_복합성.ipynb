### 제품별 분석

```python
senti_per_product = pd.DataFrame(columns=['product_name', 'sooth','moisture','cost','wrinkle','oil','whiten'])
```

```python
for p in product_list:
    df_= df.loc[df['product_name']== str(p)]
    so_li = [0,0]
    mo_li = [0,0]
    co_li = [0,0]
    wr_li = [0,0]
    oi_li = [0,0]
    wh_li = [0,0]
    
    for i in df_.loc[df_['aspect'] == '보습 및 수분감']['predict']:
        mo_li[i] += 1
    
    for i in df_.loc[df_['aspect'] == '피부진정']['predict']:
        so_li[i] += 1
    
    for i in df_.loc[df_['aspect'] == '가성비']['predict']:
        co_li[i] += 1
        
    for i in df_.loc[df_['aspect'] == '주름']['predict']:
        wr_li[i] += 1
    
    for i in df_.loc[df_['aspect'] == '유분기']['predict']:
        oi_li[i] += 1
    
    for i in df_.loc[df_['aspect'] == '미백']['predict']:
        wh_li[i] += 1
        
    new_data = {
        'product_name': p,
        'sooth': so_li,
        'moisture': mo_li,
        'cost': co_li,
        'wrinkle' : wr_li,
        'oil': oi_li,
        'whiten': wh_li
    }
    
    senti_per_product = senti_per_product.append(new_data, ignore_index=True)
```

```python
senti_per_product
```

```python
senti_per_product.to_csv('senti_product_combi.csv', index = False, columns = ['product_name', 'sooth','moisture','cost','wrinkle','oil','whiten'])
```

### 속성별 긍정 비율(%)

```python
senti_percent = pd.DataFrame(columns=['product_name', 'sooth','moisture','cost','wrinkle','oil','whiten'])
```

```python
for i in range(len(senti_per_product)):
    s_per=0
    m_per=0
    c_per=0
    w_per=0
    o_per=0
    wh_per = 0
    
    if sum(senti_per_product['sooth'][i]) != 0:
        s_per=round(senti_per_product['sooth'][i][1]/sum(senti_per_product['sooth'][i]), 4)
    
    if sum(senti_per_product['moisture'][i]) != 0 :
        m_per=round(senti_per_product['moisture'][i][1]/sum(senti_per_product['moisture'][i]),4)
    
    if sum(senti_per_product['cost'][i]) != 0:
        c_per=round(senti_per_product['cost'][i][1]/sum(senti_per_product['cost'][i]),4)
    
    if sum(senti_per_product['wrinkle'][i]) != 0:
        w_per=round(senti_per_product['wrinkle'][i][1]/sum(senti_per_product['wrinkle'][i]),4)
    
    if sum(senti_per_product['oil'][i]) != 0:
        o_per=round(senti_per_product['oil'][i][1]/sum(senti_per_product['oil'][i]),4)
    
    if sum(senti_per_product['whiten'][i]) != 0:
        wh_per=round(senti_per_product['whiten'][i][1]/sum(senti_per_product['whiten'][i]),4)
    
    new_data = {
        'product_name': senti_per_product['product_name'][i],
        'sooth': s_per*100,
        'moisture': m_per*100,
        'cost': c_per*100,
        'wrinkle' : w_per*100,
        'oil': o_per*100,
        'whiten': wh_per*100,
    }
    
    senti_percent = senti_percent.append(new_data, ignore_index=True)
```

```python
senti_percent.head(10)
```

```python
senti_percent.to_csv('senti_pos_percent_combi.csv', index = False, columns = ['product_name', 'sooth','moisture','cost','wrinkle','oil','whiten'])
```

### 속성별 언급 비율

```python
senti_percent2 = pd.DataFrame(columns=['product_name', 'sooth','moisture','cost','wrinkle','oil','whiten'])
```

```python
for i in range(len(senti_per_product)):
    s_per=0
    m_per=0
    c_per=0
    w_per=0
    o_per=0
    wh_per = 0
    #표현방법에 대해 고민중....
    total = sum(senti_per_product['sooth'][i])+sum(senti_per_product['moisture'][i])+sum(senti_per_product['cost'][i])+sum(senti_per_product['wrinkle'][i])+sum(senti_per_product['oil'][i])+sum(senti_per_product['whiten'][i])
    
    s_per=round(sum(senti_per_product['sooth'][i])/total, 4)
    m_per=round(sum(senti_per_product['moisture'][i])/total,4)
    c_per=round(sum(senti_per_product['cost'][i])/total,4)
    w_per=round(sum(senti_per_product['wrinkle'][i])/total,4)
    o_per=round(sum(senti_per_product['oil'][i])/total,4)
    wh_per=round(sum(senti_per_product['whiten'][i])/total,4)
    
    new_data = {
        'product_name': senti_per_product['product_name'][i],
        'sooth': s_per*100,
        'moisture': m_per*100,
        'cost': c_per*100,
        'wrinkle' : w_per*100,
        'oil': o_per*100,
        'whiten': wh_per*100,
    }
    
    senti_percent2 = senti_percent2.append(new_data, ignore_index=True)
```
