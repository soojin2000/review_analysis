```python
def split_review(token_review): #토큰화된 1차원 list
    topic1_keyword = ['좁쌀', '뾰루지', '흉터', '좁쌀여드름', '자국', '붉은기', '홍조', '트러블', '피지', '진정효과', '여드름', '피부염', '시어버터', '진정', '재생', '쿨링', '장벽', '판테놀', '티트리', '피부장벽']
    topic2_keyword = ['수분감', '보호막', '유수분', '보습', '속건조', '보습감', '수분', '쿨링감', '보습력', '흡수력', '지속력', '속당김', '속건조']
    topic3_keyword = ['세안', '세수', '화장전', '화장', '당김', '밀림', '메이크업', '아침저녁', '베이스', '파운데이션', '쿠션', '아침', '밀착', '기초']
    topic4_keyword = ['눈가', '팔자', '입가', '탄력', '주름', '콜라겐','레티놀']
    topic5_keyword = ['기미', '비타민', '미백']
    topic6_keyword = ['가성비', '가격대', '할인', '세일', '용량', '가격대비', '대용량', '쿠폰', '행사','가격','구성','사이즈']
    split_list = []
    i = 0
    start = 0
    while i < len(token_review):
        if token_review[i][0] in topic1_keyword or token_review[i][0] in topic2_keyword or token_review[i][0] in topic3_keyword or token_review[i][0] in topic4_keyword or token_review[i][0] in topic5_keyword or token_review[i][0] in topic6_keyword: #속성어 찾음
            split_list.append(token_review[start:find_pattern(token_review,start,i)+1]) #속성문장으로 선택
            start = find_pattern(token_review,start,i)+1 
            i=start #선택된 문장 뒤 부터 시작
        else:
            i+=1
    if len(split_list) == 0:
        split_list = [token_review]
            
    return split_list #2차원 리스트
```
