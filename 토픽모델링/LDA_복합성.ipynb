```python
def stop_nouns(sentence): 
    """ 단어의 길이가 2이상인  불용어 사전에 없는 단어를 반환한다. """
    
    # 불용어 사전
    f = open("불용어사전.txt", 'r', encoding='UTF8')
    lines = f.readlines()
    stopwords = []
    for line in lines:
        line = line.replace('\n', '')
        if len(line) > 1:
            stopwords.append(line)
            f.close()
   # 불용어 제거와 일반명사, 고유명사 추출 
    nouns = [word for word in sentence if len(word) > 1 and word not in stopwords] 
    return nouns
```

```python
from tqdm import tqdm

def dic_filter(data):
    dic_list = []
    corpus_list = [] 
    for i in tqdm([0.4,  0.5, 0.6]):
        for j in [50]:
            dictionary = gensim.corpora.Dictionary(data)
            dictionary.filter_extremes(no_below=j, no_above=i)
            bow_corpus = [dictionary.doc2bow(doc) for doc in data]
            tfidf = models.TfidfModel(bow_corpus)
            corpus_tfidf = tfidf[bow_corpus]
            dic_list.append(dictionary)
            corpus_list.append(corpus_tfidf)

    return dic_list, corpus_list
```

```python
dic_list, corpus_list = dic_filter(combination_skin_type)
```

```python
from gensim.models import LdaModel, CoherenceModel
def compute_coherence_values(combi_data,corpus, dictionary, k, c,a,b):    
    lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=dictionary,
                                           num_topics=k, 
                                           random_state=100,
                                           iterations=100,
                                           chunksize=c,
                                           alpha = a,
                                           eta = b,
                                           passes=20)
    
    coherence_model_lda = CoherenceModel(model=lda_model, texts=combi_data, dictionary=dictionary, coherence='c_v')
    
    return coherence_model_lda.get_coherence()
```

```python
import tqdm
import time
grid = {}

# Topics range
min_topics = 4
max_topics = 10
step_size = 1
topics_range = range(min_topics, max_topics, step_size)

# chunk parameter
chunk_range = [1265]

# alpha range
alpha_range = [0.1, 0.4, 0.7]

# eta range
eta_range = [0.1, 0.4, 0.7]

# Validation sets
num_of_docs = len(corpus_list[0])
corpus_sets = corpus_list
dictionary_sets = dic_list
corpus_title = ['corpus_{0}'.format(i) for i in range(1,4)]
model_results = {'Corpus_Set': [],
                 'Topics': [],
                 'Chunk' : [],
                 'Alpha' : [],
                 'Eta' : [],
                 'Coherence': []
                }
# Can take a long time to run
if 1 == 1:
   
    # iterate through validation corpuses
    for i in tqdm.tqdm((range(len(corpus_sets)))):
        time.sleep(0.25)
        # iterate through number of topics
        for k in topics_range:
            # iterate through alpha values
            for c in chunk_range:
                for a in alpha_range:
                    for b in eta_range:
                        cv = compute_coherence_values(combination_skin_type, corpus=corpus_sets[i], dictionary=dictionary_sets[i], 
                                                  k=k, c=c, a=a, b=b)
                        # Save the model results
                        model_results['Corpus_Set'].append(corpus_title[i])
                        model_results['Topics'].append(k)
                        model_results['Chunk'].append(c)
                        model_results['Alpha'].append(a)
                        model_results['Eta'].append(b)
                        model_results['Coherence'].append(cv)
                    
```

```python
pd.DataFrame(model_results).to_csv('cr_lda_tuning_results.csv', index=False)
```
